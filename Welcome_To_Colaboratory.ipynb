## Data science
</div>

With Colab you can harness the full power of popular Python libraries to analyze and visualize data. The code cell below uses **numpy** to generate some random data, and uses **matplotlib** to visualize it. To edit the code, just click the cell and start editing.

```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```

```
import pandas as pd
import numpy as np
```

```
from google.colab import drive
drive.mount('train.csv')
```

```
data = pd.read_csv('/train.csv')
data.isnull().sum()
```

```
for col in data.select_dtypes(include=['int64']):
    if data[col].isnull().sum() > 0:
        # Impute missing values with mean
        data[col] = data[col].fillna(data[col].mean())
data.columns
```

```
import matplotlib.pyplot as plt
mean_sale_price = data.groupby('LotArea')['SalePrice'].mean()

# Create the bar chart
plt.bar(mean_sale_price.index, mean_sale_price.values)
plt.xlabel('LotArea (bin)')
plt.ylabel('Mean Sale Price')
plt.title('Mean Sale Price by LotArea')
plt.show()
```

```
data.describe()
```

```
data.dtypes
```

```
# One-hot encoding for categorical features
categorical_features = ['Neighborhood']
for feature in categorical_features:
    data = pd.get_dummies(data, columns=[feature])
```

```
# Standardize numerical features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numerical_features = ['LotArea', 'GrLivArea', 'OverallQual', 'OverallCond']
data[numerical_features] = scaler.fit_transform(data[numerical_features])
```

```
data.to_csv('cleaned_house_prices_data.csv', index=False)
```

```
plt.figure(figsize=(10, 6))
plt.scatter(data['LotArea'], data['SalePrice'], alpha=0.5)
plt.xlabel('LotArea')
plt.ylabel('SalePrice')
plt.title('Sale Price vs. LotArea')
plt.grid(True)
plt.show()
```

```
plt.figure(figsize=(10, 6))
plt.scatter(data['YrSold'], data['SalePrice'], alpha=0.5)
plt.xlabel('YrSold')
plt.ylabel('SalePrice')
plt.title('Sale Price vs. Year Sold')
plt.grid(True)
plt.show()
```

```
mean_sale_price_by_zoning = data.groupby('MSZoning')['SalePrice'].mean()
plt.figure(figsize=(10, 6))
plt.bar(mean_sale_price_by_zoning.index, mean_sale_price_by_zoning.values)
plt.xlabel('MSZoning')
plt.ylabel('Mean Sale Price')
plt.title('Mean Sale Price by MSZoning')
plt.show()
```

```
import pandas as pd
from sklearn.linear_model import LinearRegression

# Separate features and target variable
X = data[['LotArea', 'OverallQual', 'GrLivArea']]
y = data['SalePrice']

# Create and train the linear regression model
model = LinearRegression()
model.fit(X, y)

# Print the model coefficients
print(model.coef_)

# Make predictions
predicted_sale_prices = model.predict(X)

# Evaluate the model performance
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y, predicted_sale_prices)
r2 = r2_score(y, predicted_sale_prices)

print('Mean Squared Error:', mse)
print('R-squared:', r2)
```

```
import matplotlib.pyplot as plt

# Extract the coefficients
coefficients = model.coef_

# Define the feature names
feature_names = ['LotArea', 'OverallQual', 'GrLivArea']

# Create the visualization
plt.figure(figsize=(10, 6))
for i, feature in enumerate(feature_names):
    plt.bar(feature, coefficients[i], label=feature)

plt.xlabel('Feature')
plt.ylabel('Coefficient')
plt.title('Impact of Features on Sale Price')
plt.legend()
plt.grid(True)
plt.show()
```
